```python
import pandas as pd
import matplotlib.pyplot as plt
#go to jupyterhub and upload your file onto jupyter
#We have 3 clusters
arr = pd.read_csv("ClusterDs1.csv")

plt.scatter("x","y",data = arr)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
2 and 3 and 4
#We have 3 clusters above
#So we have an idea of how our data should look like grouped
#Now we will use the kmeans algorithm to find this pattern

from sklearn.cluster import KMeans
km = KMeans(
n_clusters=3, 
init='random', 
n_init=10,
max_iter=300,
random_state=0 
)
y_km = km.fit_predict(arr)

plt.scatter(
arr.iloc[y_km == 0,0], arr.iloc[y_km  == 0,1], 
s=50, c='lightgreen',
marker='s', edgecolor='black',
label='cluster 1'
)

plt.scatter(
arr.iloc[y_km  == 1,0], arr.iloc[y_km  == 1,1],
s=50, c='orange',
marker='o', edgecolor='black',
label='cluster 2'
)

plt.scatter(
arr.iloc[y_km  == 2,0], arr.iloc[y_km == 2,1],
s=50, c='lightblue',
marker='v', edgecolor='black',
label='cluster 3'
)
#iloc gets the location of the point
#gets the x and then the y coordinates by the row index and columns for the listed  in the csv [index,column]]
```

```python
#5


import pandas as pd
import matplotlib.pyplot as plt
#go to jupyterhub and upload your file onto jupyter
#We have 3 clusters
CSV = pd.read_csv("ClusterDs2.csv")

plt.scatter("x","y",data = CSV)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

from sklearn.cluster import KMeans
km = KMeans(
n_clusters=3, 
init='random', 
n_init=10,
max_iter=300,
random_state=0 
)
o = km.fit_predict(CSV)

plt.scatter(
CSV.iloc[o == 0,0], CSV.iloc[o  == 0,1], 
s=50, c='lightgreen',
marker='s', edgecolor='black',
label='cluster1'
)

plt.scatter(
CSV.iloc[o  == 1,0], CSV.iloc[o == 1,1],
s=50, c='orange',
marker='o', edgecolor='black',
label='cluster2'
)

plt.scatter(
CSV.iloc[o  == 2,0], CSV.iloc[o == 2,1],
s=50, c='lightblue',
marker='v', edgecolor='black',
label='cluster3'
)

plt.scatter(
km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
s=250, marker='*',
c='red', edgecolor='black',
label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()

#km - is the Kmeans algorithm
```
